## Exercício 1

Neste exercício era pedido quantos dígitos tinha um dado número. Pelo testes, podia confirmar-se que era preciso considerar o caso do número ser negativo.

Várias soluções converteram o número numa string para contar quantas 'letras' tinha. Em Python conseguimos fazer isso facilmente. Mas uma forma mais geral seria usar a divisão inteira para ir retirando um dígito de cada vez, até se chegar ao ponto do quociente ser zero.

Esta seria a abordagem recursiva, que funciona igualmente com iteração.

```
def contarDigitos(n):
  if n<0:
    n = -n

  digitos = 1 if n==0 else 0  # dica: operador que faz as vezes do if-then-else
  while n>0:
    digitos = digitos + 1
    n = n // 10  
  return digitos
```

Existem respostas certas que mas demonstram alguma desorganização.

Neste exemplo, há um ciclo a fazer de condicional. Se começamos a usar outros comandos na vez dos comandos apropriados, o nosso código começa a tornar-se confuso e de díficil gestão,

```
def contarDigitos(n):
    string=(str(n))
    if n==0:
        return 1
    while n<1:
        elementos=len(string)
        return elementos -1
    return len(string)
```

O seguinte exemplo esconde o return no meio do ciclo, para além de misturar uma abordagem com divisão inteira e uso de *strings*.

```
def contarDigitos(n):
    index = 1
    f = abs(n)
    for i in range(0,len(str(f))):
        if(f//10 == 0):
            return index
        else:
            index +=1
            f/=10
```

Evitem este tipo de código. 

A seguinte solução não foi capaz de encontrar um padrão para o código repetido:

```
def contarDigitos(n):
    v=str(n)
    soma=0
    for x in v:
      if x=="0" or x=="1" or x=="2" or x=="3" or x=="4" or x=="5" or x=="6" or x=="7" or x=="8" or x=="9":
        soma+=1
    return soma
```

Estando a considerar letras, uma melhor solução seria,

```
def contarDigitos(n):
    v=str(n)
    soma=0
    for x in v:
      if x in "0123456789":
        soma+=1
    return soma
```

Mas de qualquer forma, o que o programa está a fazer é a função `len`, logo:

```
def contarDigitos(n):
    nDigitos = len(str(n))
    if n<0:
      nDigitos = nDigitos - 1
    return nDigitos  
```

Ou mesmo:

```
def contarDigitos(n): 
  return len(str(abs(n)))
```

Claro que estas soluções não têm ciclos. Daí a abordagem mais adequada ser a da divisão inteira.

## Exercício 3

Neste exercício dos sinais `+,-` bastava comparar se os respetivos indíces eram iguais (colocando esse valor na string resultado) ou diferentes (colocando um zero na string resultado).

Como queremos comparar os elementos do mesmo índice, podemos usar a função `zip`.

```
def neutralizarSinais(s1, s2):
  res = ''
  for a,b in zip(s1,s2):
    if a == b:
      res = res + a
    else:
      res = res + '0'
  return res
```

Houve soluções corretas mas que não intuiram este padrão, tornando-se desnecessariamente complicadas, como por exemplo,

```
def neutralizarSinais(s1, s2):
  y=""
  for x in range(0,len(s1)):
    if s1[x]=="+" and s2[x]=="-":
      y+="0"
    elif s1[x]=="-" and s2[x]=="+":
      y+="0"
    elif s1[x]=="+" and s2[x]=="+":
      y+="+"
    elif s1[x]=="-" and s2[x]=="-":
      y+="-"
  return y
```

## Exercício 4


Aqui era pedido uma lista das sucessivas acumulações de somar os primeiros n elementos.

Temos duas variáveis, uma que nos dá o elemento atual (podemos usar um `for`) e outra que acumula a soma.

Uma resposta tradicional e eficaz:

```
def somaAcc(lista):
    soma = 0
    result = []
    
    for elem in lista:
        soma = soma + elem
        result.append(soma)
        
    return result
```

A solução seguinte tem o problema de não usar uma variável acumuladora. Isto faz que estejamos sempre a criar, em cada iteração, uma sublista com os elementos a somar.

```
def somaAcc1(lista):
  if len(lista)==0:   # não há necessidade destes casos especiais
    return []
  elif len(lista)==1:
    return lista
  else:
    resultado=[]
    for n in range(0,len(lista)):
      listaSoma=sum((lista)[:n])  # criação de uma sublista desnecessária
      resultado.append(lista[n]+listaSoma)
    return resultado
```

Se a lista a somar fosse grande, estas sublistas iam começar a pesar na eficácia da função. Vejamos um exemplo,

```
listaMedia = list(range(10000))
```

```
%%timeit
somaAcc(listaMedia)
```

```
%%timeit
somaAcc1(listaMedia)
```

Demorou 3600x mais tempo para uma lista só com dez mil elementos!

Este tipo de operação de acumular resultados de uma operação é generalizável. Podemos dar como parâmetro `f` o tipo de operação (para não ser sempre uma soma).

Para além disso temos de dar o elemento neutro de `f` para que a variável acumuladora possa ser inicializada:


```
def scan(f, neutro, lista):
    acc = neutro
    result = []
    
    for elem in lista:
        acc = f(elem,acc)
        result.append(acc)
        
    return result
```

Como exercício extra, tentem perceber os seguintes exemplos,

```
soma    = lambda x,acc : x+acc
produto = lambda x,acc : x*acc
maximo  = lambda x,acc : x if x>acc else acc

lista = [4,3,7,1,3,9,6]

print(scan(soma,    0,        lista))
print(scan(produto, 1,        lista))
print(scan(maximo,  lista[0], lista))
```
