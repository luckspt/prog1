## Tuplos

O tipo **tuplo** é inspirado no tuplo matemático, corresponde uma sequência fixa de valores, ou **componentes**, de um ou mais tipos.

A sintaxe para definir um tuplo é usar parênteses, separando os valores de cada componente por vírgulas.

Um exemplo: estamos todos habituados a definir um ponto a três dimensões como um triplo de coordenadas. Ora um triplo é um tuplo com três componentes.

```
ponto3D = (1.0, 2.0, 3.0)

print(ponto3D)
```

O número de componentes de um tuplo pode variar com o que queremos representar.

```
umNome = ("João", "Neto")

print(umNome)
```

Se quisermos definir um tuplo com apenas uma componente temos de usar a seguinte sintaxe:

```
singleton = ("estou sozinho", )

print(singleton)
```

Sendo ainda possível criar o tuplo vazio:

```
tuploVazio = tuple() 
```

A função `tuple` pode ser usada para criar tuplos a partir de tipos estruturados. Um exemplo:

```
stringParaTuplo = tuple("componentes")

print(stringParaTuplo)
```

Como referido no início, as componentes podem ser de tipos distintos:

```
# exemplo de um registo com nome, apelido, identificador, altura e se é docente
registoPessoa = ("João", "Neto", 41455, 1.72, True)
```

Podemos usar o operador de atribuição para aceder às componentes do tuplo,

```
nome, apelido, id, altura, eProfessor = registoPessoa

print( "{0} tem {1}m de altura".format(nome, altura) )
```

Dado um tuplo podemos aceder às componentes que quisermos através da sintaxe `tuplo[i]` sendo `i` chamado de **índice**, isto é, a posição da componente no tuplo. Por convenção, a primeira componente de um tuplo tem índice zero.

```
print( "{0} tem {1}m de altura".format(registoPessoa[0], registoPessoa[3]) )
```

Tipos estruturados aos quais podemos aceder via índices chamamos **tipos indexados**.

Outro tipo indexado são as *strings*, podemos usar o mesmo mecanismo para aceder a certos caracteres da frase:

```
umaString = "Olá Mundo"

print(umaString[2])
```

O Python é bastante uniforme como trata os seus tipos estruturados. Este é um exemplo. Todos os tipos estruturados que sejam indexados usam esta sintaxe dos parênteses retos para aceder à informação.



---


Os operadores de adição e multiplicação também estão definidos para tuplos. 

A expressão `tuplo1 + tuplo2` juntas as componentes de ambos os tuplos num tuplo resultado. A expressão `tuplo * n` repete as componentes do tuplo `n` vezes:

```
print(singleton + registoPessoa)

print(singleton * 3)   # transforma o singleton num triplo
```

Os tuplos têm algumas funções associadas:

+ `index` devolve o primeiro índice onde está uma dada componente

+ `count` conta quantos componentes são iguais a um dado valor

+ `len` devolve o número de componentes

```
oMeuTuplo = (1, 2, 3, 1, 1)

print(oMeuTuplo.index(3))
print(oMeuTuplo.count(1))

print(len(oMeuTuplo))
```

Reparem que a sintaxe para executarmos estas funções nem sempre foi a mesma. Para o `index` e `count` usámos um ponto para separar o tuplo da função. Nesta altura do campeonato ainda é cedo para explicar o porquê desta diferença.

Existem outras funções que o Python aceita executar se todas as componentes de um tuplo forem números. Alguns exemplos:

```
print(sum(oMeuTuplo))  # somatório das componentes
print(max(oMeuTuplo))  # devolve a componente máxima

oMeuOutroTuplo = (1, 2, 4, 1, 1)
print(oMeuTuplo < oMeuOutroTuplo) # compara cada componente à vez, com prioridade decrescente
```

Para os tipos indexados o Python tem outra funcionalidade chamada de **slicing** (cortar, fatiar?).

Com *slicing* podemos rapidamente ir buscar partes de um tipo indexado.

Dada uma variável `v` de um tipo indexado, `v[i:j]` devolve todos os valores cujos índices vão de `i` (inclusive) até `j` (exclusive). Esta notação aceita também índices negativos, onde o índice `-1` representa o último elemento, `-2` o penúltimo elemento, etc.

Alguns exemplos:

```
facesDado = ("um", "dois", "três", "quatro", "cinco", "seis")

print(facesDado)
print(facesDado[1:3])   # do índice 1 (inclusive) ao índice 3 (exclusive)
print(facesDado[:2])    # desde o ínicio até ao índice 2 (exclusive)
print(facesDado[2:])    # desde o índice 2 até ao fim
print(facesDado[-3:-1]) # do antepenultimo (índice -3) ao último (índice -1) (exclusive)
print(facesDado[:-1])   # todos menos o último
print(facesDado[:])     # uma cópia do tuplo inteiro
```

Como o pessoal do Python nunca está satisfeito, ainda há um outro truque com o *slicing*. Se escreverem `a:b:c` o `c` representa incrementos. Por exemplo, `v[1:10:2]` só vai devolver os índices `1, 1+2, 1+4` etc. 

```
print(facesDado[1:5:2])
print(facesDado[1::2])
print(facesDado[:1:-1])
print(facesDado[::-1])
```

Uma última nota sobre os tuplos. Uma vez definido um tuplo não o podemos alterar. Vamos tentar para ver o que acontece:

```
umTuplo = (1, 2, 3)

umTuplo[1] = 0
```

Ocorreu um erro de execução porque tentámos alterar uma das componentes do tuplo. Um tipo cujos valores não se podem modificar designa-se por **imutável**.

Outro tipo imutável que já conhecemos são as *strings*.

```
umaString = "olá Mundo"

umaString[1] = "O"
```

Em contraste, a um tipo que podemos modificar parte do seu valor diz-se **mutável**. Vamos conhecer já a seguir um tipo estruturado que é mutável, as listas.

## Listas

As listas são, provavelmente, o tipo estruturado mais importante do Python.

Uma **lista** é um tipo indexado mutável que armazena uma sequência de valores. Sendo mutável significa que é possível alterar, durante a execução do programa, os elementos de uma lista.

Enquanto um tuplo tem um número fixo de componentes, o número de elementos guardados numa lista pode variar.

As listas sintaticamente são identificadas por parênteses retos `[]` sendo os vários valores separados por vírgulas. Apesar dos parênteses retos serem o mesmo elemento sintático que usamos para indexar tipos, o Python não se confunde.


```
umaLista = [1, 2, 3, "a"]   # podem ter valores de tipos diferentes como os tuplos

print(umaLista)
print(umaLista[0])

print('-'*10)

# O Python até se desenrasca neste exemplo. E vocês?
print([0][0])
```

Algumas funções úteis para manipulação de listas:

+ o operador `+` concatena duas listas

+ o operador `*` multiplica o conteúdo de uma lista (similar aos tuplos)

+ o operador `in` verifica se um elemento está numa lista

+ `len(lista)` devolve o número de elementos da lista

+ `lista.append(x)` junta o elemento `x` ao fim da lista

+ `lista.extend(xs)` junta os elementos do tipo estruturado `xs` ao fim da lista

+ `lista.insert(i, x)` insere o elemento `x` no índice `i` (a lista cresce um elemento)

+ `lista.remove(x)` remove da lista a primeira ocorrência do elemento `x`

+ `lista.index(x, start, end)` retorna o índice da primeira ocorrência do elemento `x` (opcionalmente) no intervalo definido pelos valores `start` até `end` (exclusive). Se `end` não for dado, a pesquisa é até ao fim da lista.

+ `lista.sort()` ordena a lista por ordem crescente *desde* que os elementos sejam comparáveis entre si.

```
umaLista.append(4)
print(umaLista)

print(10 in umaLista)  # o elemento 10 pertence à lista?

umaLista.extend([5, 10])
print(umaLista)

umaLista.extend((2, True))
print(umaLista)

umaLista.insert(1, "Olá!")
print(umaLista)

print(umaLista.index(2, 4))  # o índice do primeiro 2 a partir do índice 4
```

As listas têm o mesmo mecanismo de *slicing* que os tuplos.

```
quadrados = [1, 4, 9, 16, 25, 36]

print(quadrados[3:])
```

As listas têm uma atribuição múltipla que nos permite facilmente separar o primeiro elemento da lista do resto da lista:

```
lista = [1,2,3,4,5,6]

primeiro, *resto = lista

print(primeiro)
print(resto)
```

ou separar o último elemento:

```
*inicio, ultimo = lista

print(inicio)
print(ultimo)
```

As listas são tão flexíveis que até podemos ter listas de listas

```
listaDeListas = [ [3,4,5], [6,8,9] ]
```

### Listas por Compreensão

As listas por compreensão é uma funcionalidade do Python para criar, de forma concisa, listas baseadas em outros valores estruturados.

A ideia da notação que vamos aprender inspira-se na notação usada na Matemática para definir conjuntos.

Por exemplo, este conjunto representa todos os quadrados de números naturais menores que 20:

$$\{ x^2 : x \in \mathbb{N^+}, x < 20 \}$$

como podemos criar uma lista com os mesmos elementos?

```
print( [ x**2 for x in range(1, 20) ] )
```

Vamos olhar com atenção para este comando.

<center><code>  <font size="+1">
[ <font color='red'>x**2</font> <font color='blue'>for</font>  <font color='red'>x</font> <font color='blue'>in</font>  <font color='purple'>range(1, 20)</font> ]
</font></code></center>

+ A expressão inicial `x**2` corresponde à expressão comum partilhada pelos elementos da lista final. Neste caso serão quadrados de um certo número `x`.

+ O que se escreve a seguir à palavra reservada <font color='blue'>for</font> define como os vários `x` vão ser gerados.

+ <font color='red'>x</font> <font color='blue'>in</font>  <font color='purple'>range(1, 20)</font> define como se geram os valores `x`. Neste caso usamos uma função especial do Python `range` que gera todos os números de 1 a 20 (exclusive).

* À variável `x` que vai recebendo os valores do gerador costumamos chamar de **variável de progresso**.



Vejamos algumas variações:

```
print( [ x**3 for x in range(1,20)] )  # mudar a expressão de x^2 para x^3

print( [ x**2 for x in range(5,10)] )  # mudar o range

print( [ x**2 for x in range(10) ] )   # range(n) gera 0,1,...,n-1

print( [ x for x in range(20,1,-2) ] ) # range(a,b,c) gera a,...,b com acrescentos c
```

Se o valor da variável de progresso não for relevante para a construção dos valores da lista, é costume nomear a variável de progresso por `_` :

```
print( [ 1 for _ in range(10) ] )  # lista de dez uns
```

A expressão inicial não precisa ser numérica. Alguns exemplos de outros tipos:

```
print( [ x > 3 for x in range(6) ] )

print( [ (x, x > 3) for x in range(6) ] )

print( [ [x, x+1] for x in range(4) ] )
```

As listas por compreensão permitem igualmente eliminar alguns valores que não queiramos.

Um **filtro** é uma expressão booleana que tem de ser satisfeita para que o  elemento em questão faça parte da lista final. Um filtro surge depois da palavra reservada `if`.

No exemplo seguinte, vamos devolver todos os tuplos possíveis $(x,y)$, onde $1 \leq x,y \leq 3$, desde que $x \neq y$:

```
[ (x, y) for x in range(1,4) 
         for y in range(1,4) 
         if x != y ]
```

### Interlúdio

É necessário falarmos um pouco sobre o que se passa nos bastidores, nomeadamente no que toca às variáveis que vamos criando ao longo da execução do programa.

Quando o Python executa o comando `x = 1` o que acontece?

```
#@title 
from graphviz import Digraph

styles = {
    'ref': {'shape': 'doublecircle'},
    'val': {'shape': 'ellipse'}
}

g = Digraph(engine="neato")
g.node('X', 'X',  styles['ref'], pos='0,2!')
g.node('V1', '1', styles['val'], pos='0,1!')
g.edge('X', 'V1')
g
```

É reservada uma posição de memória que fica associada à variável `x`. Esta posição de memória *guarda o endereço* onde está o valor inteiro 1. Neste contexto, diz-se que `x` refere o valor 1. No diagrama a variável é indicada pela dupla borda. O círculo simples com o 1 representa a posição de memória onde o valor 1 se encontra.

Uma **referência** é a associação entre uma variável e o valor que existe no endereço guardado pela variável. Esta associação é aqui representada pela seta.

Nesta situação, como irá decorrer a execução do novo comando `x = x+1`?

1. O Python ao avaliar a expressão `x+1` deteta o nome da variável `x`. O que faz é ir buscar, via a respetiva referência, o valor associado `1`

1. O valor final da expressão é `2` o que resulta na reserva de nova memória para guardar o valor inteiro `2`

1. O endereço onde se encontra o valor `2` é enviado para a variável que está à esquerda do comando de atribuição (neste caso é novamente a variável `x`, mas podia ser outra variável qualquer) criando uma nova referência entre `x` e `2`.

Assim, se executarmos `x = x+1` ficamos com

```
#@title 
g = Digraph(engine="neato")
g.node('X', 'X',  styles['ref'], pos='0,2!')
g.node('V1', '1', styles['val'], pos='0,1!')
g.node('V2', '2', styles['val'], pos='1,1!')
g.edge('X', 'V2')
g
```

O valor 1 fica inacessível e, se calhar, já não faz falta. A gestão de memória está a cargo do Python, ele decidirá apagar valores inúteis quando achar por bem. Designa-se a este processo de limpeza de memória por **garbage collection** e, felizmente para a nossa sanidade mental, não temos de nos preocupar com isso.

Mas há outras coisas que temos de nos preocupar!

```
l1 = [1,2,3]
```

Ao executar o comando acima, a organização de memória é dada pelo diagrama seguinte. 

```
#@title 
g = Digraph(engine="neato")
g.node('ref1', 'l1',  styles['ref'], pos='0,3!')
g.node('list1', ' ', styles['val'], pos='0,2!')

g.node('val1', '1', styles['val'], pos='0,1!')
g.node('val2', '2', styles['val'], pos='1,1!')
g.node('val3', '3', styles['val'], pos='2,1!')

g.edge('ref1', 'list1')

g.edge('list1', 'val1')
g.edge('list1', 'val2')
g.edge('list1', 'val3')

g
```

Notem que uma lista tem uma estrutura interna, onde se incluem um conjunto de referências para os valores da lista. Não conhecemos detalhes desta estrutura interna, mas isso não é importante aqui.

Outro pormenor é que a lista propriamente dita não tem um nome, apenas a variável `l1`. Nós criámos a variável explicitamente mas pedimos ao Python para nos arranjar uma lista nova. Os valores criados pela avaliação de expressões são anónimos. A única forma que temos de os aceder é através das variáveis que usamos.

```
l2 = l1
l2.append(4)

print(l1)
```

Aqui está um comportamento inesperado! Adicionámos um novo elemento na lista l2, mas ele apareceu como por magia na lista l1. O que aconteceu?

Isto ocorre porque não há duas listas, há é duas referências para listas. E ambas as variáveis referenciam a mesma lista anónima!

```
#@title 
g = Digraph(engine="neato")
g.node('ref1', 'l1',  styles['ref'], pos='0,3!')
g.node('ref2', 'l2',  styles['ref'], pos='2,3!')
g.node('list1', ' ', styles['val'], pos='1,2!')

g.node('val1', '1', styles['val'], pos='0,1!')
g.node('val2', '2', styles['val'], pos='1,1!')
g.node('val3', '3', styles['val'], pos='2,1!')
g.node('val4', '4', styles['val'], pos='3,1!')

g.edge('ref1', 'list1')
g.edge('ref2', 'list1')

g.edge('list1', 'val1')
g.edge('list1', 'val2')
g.edge('list1', 'val3')
g.edge('list1', 'val4')

g
```

O operador `is` ajuda-nos a verificar se duas variáveis referenciam o mesmo valor.

```
print(l1 is l2)
```

A função `id` mostra o endereço associado à referência (os detalhes são mais complicados mas estão fora do âmbito da disciplina). O nome da função é uma abreviatura de **identidade**.

```
print(id(l1))     # l1 e l2 têm a mesma referência, logo
print(id(l2))     #  as suas identidades são iguais

print(id(l1[0]))  # onde está o primeiro elemento da lista referenciada por l1
print(id(l2[0]))  # onde está o primeiro elemento da lista referenciada por l2
```

Então como é que criamos uma segunda lista para a poder alterar sem que `l1` se aperceba? Temos de dizer ao Python que queremos uma cópia da lista original.

```
l1 = [1,2,3]
l2 = list(l1) # devolve uma cópia da lista. Em alternativa l2 = l1[:] também serve

print("l1 is l2? {0}".format(l1 is l2))
print("id lista1 = {0}, id lista2 {1}".format(id(l1), id(l2)))

l2.append(4)

print(l1)
```

Muito melhor :-)  

Esta é a memória após a execução deste programa:

```
#@title 
g = Digraph(engine="neato")
g.node('ref1', 'l1',  styles['ref'], pos='0,3!')
g.node('ref2', 'l2',  styles['ref'], pos='2,3!')
g.node('list1', '', styles['val'], pos='0,2!')
g.node('list2', '', styles['val'], pos='2,2!')

g.node('val1', '1', styles['val'], pos='0,1!')
g.node('val2', '2', styles['val'], pos='1,1!')
g.node('val3', '3', styles['val'], pos='2,1!')
g.node('val4', '4', styles['val'], pos='3,1!')

g.edge('ref1', 'list1')
g.edge('ref2', 'list2')

g.edge('list1', 'val1')
g.edge('list1', 'val2')
g.edge('list1', 'val3')

g.edge('list2', 'val1')
g.edge('list2', 'val2')
g.edge('list2', 'val3')
g.edge('list2', 'val4')

g
```

Mesmo assim, esta técnica ainda pode levantar problemas se os valores da lista forem, eles próprios, listas. 

As cópias efetuadas apenas duplicaram a estrutura da lista mas não os valores referenciados pela lista inicial. A isto chama-se uma **shallow copy** (uma cópia superficial). Quando os valores guardados pelas listas são imutáveis não há consequências em haver partilha de informação. Mas se os valores em si também forem mutáveis temos problemas...

Vamos ver um exemplo:

```
l1 = [1, 2, [3, 4]]
l2 = l1[:]             # cópia superficial
l2[2][1] = 666

print(l1)             # um problema diabólico
```

Eis a memória quando o programa acima terminou:

```
#@title 
g = Digraph(engine="neato")
g.node('ref1', 'l1',  styles['ref'], pos='0,3!')
g.node('ref2', 'l2',  styles['ref'], pos='2,3!')
g.node('list1', '', styles['val'], pos='0,2!')
g.node('list2', '', styles['val'], pos='2,2!')

g.node('val1', '1', styles['val'], pos='0,1!')
g.node('val2', '2', styles['val'], pos='1,1!')
g.node('val3', '', styles['val'], pos='2,1!')
g.node('val3a', '3', styles['val'], pos='1.5,0!')
g.node('val3b', '666', styles['val'], pos='2.5,0!')

g.edge('ref1', 'list1')
g.edge('ref2', 'list2')

g.edge('list1', 'val1')
g.edge('list1', 'val2')
g.edge('list1', 'val3')

g.edge('list2', 'val1')
g.edge('list2', 'val2')
g.edge('list2', 'val3')

g.edge('val3', 'val3a')
g.edge('val3', 'val3b')

g
```

Se quisermos efetuar uma cópia de tudo temos de usar a função `deepcopy` do módulo `copy`. 

Esta funcionalidade, porém, tem de ser usada com cuidado. Se a estrutura que queremos copiar estiver associada a muita informação, uma cópia profunda pode gastar imensos recursos.

```
import copy

l1 = [1, 2, [3, 4]]
l2 = copy.deepcopy(l1) # cópia profunda
l2[2][1] = 666

print(l1)
print(l2)
```

Com o deep copy esta é a organização da memória no fim das atribuições:

Um outro detalhe. Existe uma diferença subtil, mas importante, entre o operador de igualdade `==` e o de identidade `is`. O primeiro compara os valores associados às variáveis. O segundo compara as referências, isto é, os endereços onde estão esses valores. Pensando nos grafos que vimos acima, se se trata do mesmo nó.

Um pequeno exemplo:

```
lista1 = [1,2,3]
lista2 = lista1
lista3 = list(lista1)

# todas as listas são iguais, ie, são compostas pelos mesmos elementos
print(lista1 == lista2, lista2 == lista3, lista1 == lista3)

# há duas listas distintas, logo há duas referências distintas
print(lista1 is lista2, lista2 is lista3, lista1 is lista3)
```

Em resumo, a partir da referência associada à variável temos acesso a três informações:

+ o seu valor

+ o tipo do valor (a variável não tem um tipo, apenas o valor. Isto é diferente noutras linguagens, como o Java ou o C)

+ a identidade do valor

```
umaVariavel = 100

print(umaVariavel, type(umaVariavel), id(umaVariavel), sep=' --- ')
```

## Estudo de um Problema

Vamos terminar este capítulo com a resolução de um problema mais elaborado. O tema é o  lançamento de dados.

Para representar o lançamento aleatório de dados temos de usar o módulo `random` que inclui funções para gerar números aleatórios. No nosso caso, vamos trabalhar com a função `randint(a,b)` que gera um inteiro entre `a` e `b` (inclusive). 

```
import random

print( [random.randint(1,6) for _ in range(10)] )  # lançamento de dez dados
```

O problema é este. Se lançarmos três dados e os somarmos, sabemos que há resultados mais prováveis que outros. Por exemplo, a soma 3 (todos os dados saíram na face um) é mais improvável de sair que a soma 10. Gostaríamos de saber as proporções dos vários valores possíveis da soma.

Este é uma pergunta típica de probabilidades. Nós aqui vamos tentar dar uma resposta aproximada através de uma simulação. O que significa simular neste contexto? 

Podemos usar o Python para realizar a experiência de lançar três dados e somá-los:

```
numDados = 3

lancamentoDados = [random.randint(1,6) for _ in range(numDados)]
experiencia = sum(lancamentoDados)

print(experiencia)
```

Se fizermos esta experiência *muitas* vezes, e contabilizarmos as várias somas que vão saindo, podemos ter uma boa ideia das proporções para cada valor da soma.

```
numExperiencias = 100000

# repetir muitas vezes: lançar três dados e somar
experiencias = [ sum([random.randint(1,6) for _ in range(numDados)]) 
                 for _ in range(numExperiencias) ]

# os resultados possíveis de somar três dados vão de 3 a 18, ie, range(3,19)
contarSomas = [ experiencias.count(resultado)  for resultado in range(3,19) ]

# transformar em proporções
proporcoes = [ soma/numExperiencias for soma in contarSomas ]

print(proporcoes)
```

Podemos criar um relatório mais informativo:

```
relatorio = "proporção de somar {0:2d} é ≈ {1:4.1f}%"

[ print( relatorio.format(resultado, 100*proporcoes[resultado-3]) ) 
  for resultado in range(3,19) ];
```

<font size="+4" color="blue;green"><b>?</b></font> <font size="+4" color="blue;green"><b>?</b></font> <font size="-1">[problema mais difícil]</font> O [Paradoxo do Aniversário](https://pt.wikipedia.org/wiki/Paradoxo_do_anivers%C3%A1rio) é um resultado contra-intuitivo da Teoria das Probabilidades que diz se tivermos 23 pessoas numa sala, há cerca de 50% de probabilidade de pelo menos duas dessas pessoas partilharem a data de aniversário. Gostaríamos de usar o Python para ter uma ideia se realmente isto é verdade.

Vamos considerar uma data de aniversário como um número inteiro entre 1 (1º de Janeiro) e 365 (31 de Dezembro). 

O que pretendemos fazer é *simular* este problema. Vamos designar por *experiência* o criar aleatoriamente um grupo de 23 datas de aniversário, e verificar se há aniversários partilhados. Se houver aniversários partilhados dizemos que a experiência foi um sucesso. 

O truque que faz funcionar a simulação é fazer muitas experiências, por exemplo 10 mil experiências. Se contabilizarmos as experiências com sucesso e dividirmos pelo número total de experiência, temos uma proporção de sucessos. Esta proporção é o resultado da simulação e uma aproximação razoável da probabilidade descrita pelo paradoxo. Será que dá mesmo um número perto dos 50%?

O programa seguinte tem exemplos de comandos úteis à resolução. Se o executarem várias vezes, vão confirmar que os resultados mudam. Isto é normal dado que estamos a usar aleatoriedade.

```
import random  # módulo para gerar números aleatórios

numPessoas = 23
numExperiencias = 10000

i = 0
qtd = 0
while i<numExperiencias:
  grupo = [ random.randint(1,365) for _ in range(numPessoas) ]
  resultado = max( [grupo.count(aniversario) for aniversario in grupo] ) 
  if resultado > 1:
    qtd+=1;
  i+=1

print("Qtd: {0}\nProb: {1:.2f}%".format(qtd, (qtd/numExperiencias) * 100))

##### fim da experiência

   # ponham aqui a vossa solução
```
